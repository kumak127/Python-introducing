## 5.5.1  setdefaultとdefaultdictによる存在しないキーの処理

```python
alp_num = {"a":1,"b":2,"c":3}
```

```python
alp_num.setdefault("d",4)
```

```python
alp_num
```

```python
alp_num.setdefault("c",5)
```

辞書のキーとして存在するか分からないけど、ない場合は特定の値をペアとして代入したいときは setdefault メソッドを使うと便利  
第一引数に探したいキーをいれ、第二引数にキーがなかった場合にペアにする値を入れる  
キーがない場合は、第二引数に渡した値とペアとして新しく辞書に代入され、すでにキーがあった場合は変更されずにすでに代入されている値を返す

```python
from collections import defaultdict
```

```python
alp_num = defaultdict(int)
```

```python
alp_num["a"] = 1
```

```python
alp_num["b"]
```

```python
alp_num
```

パッケージ collections の defaultdict は引数として関数を渡すことで、キーを指定して呼び出して値がなかった場合に自動で渡した関数の型を代入する  
(intの場合0、listの場合[ ])

```python
def idont_know():
    return "?"
```

```python
dont_understand = defaultdict(lambda :"?") # 上と同じ
```

```python
dont_understand["big_blue"] = "sea"
```

```python
dont_understand["circle_red"]
```

```python
dont_understand["small_yellow"]
```

```python
dont_understand
```

defaultdictには無名関数のlambdaを渡すことで、簡単に定義付けをすることができる

```python
word_counter = defaultdict(int)
```

```python
for word in "hifushagihonjgdvbifuhcondsiguwhao;fndvouvhasug;rhvbndjndlifuh":
    word_counter[word] += 1
```

```python
word_counter
```

defaultdictを使うことで本来エラーが起こる場合にも自動で値を入れてくれるためエラーが起こらなくなる

## 5.5.2  Counterによる要素数の計算

```python
from collections import Counter
```

```python
word = "hfidushghergbnkdjnclnjgouryuhiehfnvjdklviohrey"
```

```python
word_counter = Counter(word)
```

```python
word_counter
```

```python
type(word_counter)
```

```python
word_counter.most_common(5)
```

パッケージcollectionsの Counter 関数にシーケンスを渡すと要素が出てくる回数を Counterオブジェクト として辞書のようにまとめて返してくれる  
Counterオブジェクトのmost_commonメソッドに整数を渡すことで上位の整数分だけリストの中のタプルとして返してくれる

```python
word_counter1 = Counter("ghdishid")
word_counter2 = Counter("fnisjoie")
```

```python
word_counter1 + word_counter2 # 2つのカウンタを結合
```

```python
word_counter1 - word_counter2 # 1つめのカウンタから2つめのカウンタを引く
```

```python
word_counter1 & word_counter2 # 2つのカウンタの積集合
```

```python
word_counter1 | word_counter2 # 2つのカウンタの和集合
```

Counterオブジェクトはお互いに結合・計算することができる

## 5.5.4  スタック+キュー=デック

スタックとはLast-In-First-Outのことで、キューはFirst-In-First-Outのことを指す  
デックはスタックとキューの両方を兼ね備えたもので、先頭からでも末尾からでも選んで要素を追加・削除できることを指す

```python
from collections import deque
```

```python
dq = deque(["a","b","c"])
```

```python
type(dq)
```

```python
dq.append("d")
```

```python
dq.appendleft("s")
```

```python
list(dq)
```

```python
dq.pop()
```

```python
dq.popleft()
```

collectionsパッケージの deque 関数にシーケンスを渡すことで、先頭からでも末尾からでも要素を追加・削除することができる deque オブジェクトを生成することができる  
list型のように末尾から追加・削除するときは同じようにappend,pop,extendを使い、  
先頭から追加・削除したいときはappendleft,popleft,extendleftを使うことでできる

```python
def palindrome(word):
    from collections import deque
    dq = deque(word)
    while len(dq) > 1:
        if dq.popleft() != dq.pop():
            return False
    return True
```

```python
palindrome("a")
```

```python
palindrome("racecar")
```

```python
palindrome("halibut")
```

## 5.5.5  itertoolsによるコード構造の反復処理

```python
import itertools
```

```python
for item in itertools.chain([1,2],["a","b","c"],["d",["e","f"]]):
    print(item)
```

itertoolsモジュールにはいろいろな機能持つイテレータ関数がいくつかあり、for文のループで呼び出すことで機能できる  
itertoolsの chain 関数は引数として渡した要素をすべてつなげて1つのイテラブルであるかのように扱うことができる

```python
for item in itertools.cycle([1,2]):
    print(item,end="")
```

itertoolsモジュールの cycle 関数は引数の要素を無限に繰り返し取り出しwhileループのような処理をすることができる

```python
for item in itertools.accumulate([1,2,3,4]):
    print(item)
```

```python
for item in itertools.accumulate([1,2,3,4],lambda x,y : x*(y+1)):
    print(item)
```

itertoolsモジュールの accumulate 関数は引数に渡した要素を計算しながら値を返してくれる  
第2引数に計算内容を渡すことで指定した通りの計算をしながら返してくれる  
(上ではxが一つの前の数字で、yが現在の数字の値になる)
